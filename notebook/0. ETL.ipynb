```python
import pandas as pd 

movie = pd.read_csv('../data/movies_metadata.csv')
rating = pd.read_csv('../data/ratings_small.csv')
link = pd.read_csv('../data/links_small.csv')
```

```python
merged_1 = pd.merge(link, rating)
```

```python
merged_1['imdb_id'] = 'tt' + merged_1['imdbId'].astype(int).astype(str).str.zfill(7)
df = pd.merge(merged_1, movie, on='imdb_id', how='inner')
```
